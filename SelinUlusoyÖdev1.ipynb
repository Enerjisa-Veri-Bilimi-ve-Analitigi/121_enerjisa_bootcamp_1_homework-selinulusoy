## Soru 2

### Ekli data üzerindeki aşağıda yer alan kolonların count, mean, std, min, 25%, 50%, 75%, max değerlerini ve IQR metoduyla outlier kalan veri sayısını çıkartıp, bu değerleri sözel ve karşılaştırmalı olarak yorumlayalım.

Öncelikle verileri bir DataFrame içinde görselleştirip yorumlayalım

```python
data = pd.read_csv("data.csv")                                
df = pd.DataFrame(data) 
df.head()
```

Veri seti, iki farklı tümör çeşidi olan Malignant ve Benign hücrelerinin çekirdeklerinin özelliklerini tanımlar. Malignant kanserli hücre iken Benign kanser olmayan hücredir. Bu projede  hücre çekirdeklerinin radius (merkezden çevre üzerindeki noktalara olan mesafelerin ortalaması), area (alan) ve compactness (çevre ^ 2 / alan - 1.0) özelliklerini karşılaştırmalı olarak yorumlayacağız.

#### Benign türünün count, mean, std, min, 25%, 50%, 75%, max değerlerini bulup IQR metoduyla outlier kalan veri sayısını çıkartalım

```python
selected_columns = df[['radius_mean', 'area_mean', 'compactness_mean']]  
new_df= selected_columns.copy()

benign_df= new_df[df["diagnosis"]=="B"]

benign_df.describe()
```

```python
columns = ['radius_mean', 'area_mean', 'compactness_mean']

Q1= benign_df[columns].quantile(0.25)
Q3= benign_df[columns].quantile(0.75)
IQR = Q3 -Q1

lower_lim = Q1 - 1.5*IQR
upper_lim = Q3 + 1.5*IQR

no_outliers_benign_df = benign_df[~((benign_df[columns]<lower_lim) | (benign_df[columns]>upper_lim)).any(axis=1)]
benign_outliers = benign_df[~(benign_df[columns].isin(no_outliers_benign_df[columns]))].reset_index(drop=True)
benign_outliers.dropna(subset=['radius_mean', 'area_mean', 'compactness_mean'],inplace=True)


print(f"Outlier değerler çıktıktan sonraki tablomuz aşağıdaki gibidir. {len(benign_outliers)} satır silindi")
no_outliers_benign_df
```

#### Malignant türünün count, mean, std, min, 25%, 50%, 75%, max değerlerini bulup IQR metoduyla outlier kalan veri sayısını çıkartalım

```python
selected_columns = df[['radius_mean', 'area_mean', 'compactness_mean']]  
new_df= selected_columns.copy()

malignant_df= new_df[df["diagnosis"]=="M"]

malignant_df.describe()
```

```python
Q1= malignant_df[columns].quantile(0.25)
Q3= malignant_df[columns].quantile(0.75)
IQR = Q3 -Q1

lower_lim = Q1 - 1.5*IQR
upper_lim = Q3 + 1.5*IQR

no_outliers_malignant_df = malignant_df[~((malignant_df[columns]<lower_lim) | (malignant_df[columns]>upper_lim)).any(axis=1)]
malignant_outliers = malignant_df[~(malignant_df[columns].isin(no_outliers_malignant_df[columns]))].reset_index(drop=True)
malignant_outliers.dropna(subset=['radius_mean', 'area_mean', 'compactness_mean'],inplace=True)


print(f"Outlier değerler çıktıktan sonraki tablomuz aşağıdaki gibidir. {len(malignant_outliers)} satır silindi.")
no_outliers_malignant_df
```

İstatistiksel tabloya göre Benign ve Malignant hücreleri arasındaki birtakım farklar ayırt edilebilir. min,max ve mean değerler karşılaştırıldığında, Malignant türü hücre çekirdeklerinin yarıçap (radius_mean) ve dolayısıyla alan (area_mean) değerlerinin daha yüksek bir ortalamaya sahip olduğunu gözlemlemekteyiz. Aynı durum çekirdeğin yoğunluk (compactness_mean) değerleri için de geçerlidir. Çekirdeğin radius_mean, area_mean ve compactness_mean değerlerinin, hücrenin kanser olup olmadığını anlamada ayırt edici olduğu söylenebilir.

1)  S: Aşağıdaki istatistikler göz önüne alındığında, bir kadının pozitif mamografi sonucu varsa kanser olma olasılığı nedir?
    50 Yaşın üzerindeki kadınların yüzde birinde meme kanseri var.
    Meme kanseri olan kadınların yüzde doksanı mamogramlarda pozitif test alıyor.
    Kadınların yüzde sekizi yanlış pozitif olacak.

    P(A) = 0,01
    P (~ A) = 0,99
    P (X|A) = 0.9
    P (X |~ A) = 0,08

    (0.9 * 0.01) / ((0.9 * 0.01) + (0.08 * 0.99) = 0.10.

    Pozitif test sonucu verilen bir kadının kanser olma olasılığı % 10'dur.
